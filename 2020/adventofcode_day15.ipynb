```python
numbers = [11, 0, 1, 10, 5, 19]
for new_number in range(6, 9):
    if numbers[new_number-1] not in numbers[:-1]:
        numbers.append(0)
    else:
        numbers.append(new_number - ([i for i, val in enumerate(numbers[:-1]) if val == numbers[-1]][-1]+1))
numbers
```

```python
numbers = {11: 0, 0: 1, 1: 2, 10:3, 5: 4, 19: 5}
add_number = 0
for new_number in range(6, 30000000):
    #print(add_number)
    if add_number in numbers.keys():
        hold  = numbers[add_number]
        numbers[add_number] = new_number
        add_number = new_number - hold
    else:
        numbers[add_number] = new_number
        add_number = 0
#numbers[30000000-1]
```
